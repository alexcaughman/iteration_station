```python

#dependencies
import csv
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
#read in zip csv's
zip30312 = "Weather_Traffic_Resources/30312_weather_traffic_FINAL.csv"
zip30339 = "Weather_Traffic_Resources/30339_weather_traffic_FINAL.csv"
zip30331 = "Weather_Traffic_Resources/30331_weather_traffic_FINAL.csv"
zip30315 = "Weather_Traffic_Resources/30315_weather_traffic_FINAL.csv"
zip30309 = "Weather_Traffic_Resources/30309_weather_traffic_FINAL.csv"
zip30340 = "Weather_Traffic_Resources/30340_weather_traffic_FINAL.csv"
zip30308 = "Weather_Traffic_Resources/30308_weather_traffic_FINAL.csv"
zip30324 = "Weather_Traffic_Resources/30324_weather_traffic_FINAL.csv"
zip30342 = "Weather_Traffic_Resources/30342_weather_traffic_FINAL.csv"
zip30318 = "Weather_Traffic_Resources/30318_weather_traffic_FINAL.csv"

zip30312_df = pd.read_csv(zip30312)
zip30339_df = pd.read_csv(zip30339)
zip30331_df = pd.read_csv(zip30331)
zip30315_df = pd.read_csv(zip30315)
zip30309_df = pd.read_csv(zip30309)
zip30340_df = pd.read_csv(zip30340)
zip30308_df = pd.read_csv(zip30308)
zip30324_df = pd.read_csv(zip30324)
zip30342_df = pd.read_csv(zip30342)
zip30318_df = pd.read_csv(zip30318)

zip30312_df.head()
```

```python
#accidents with temperature over 80
HiTemp30312 = zip30312_df.loc[zip30312_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30339 = zip30339_df.loc[zip30339_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30331 = zip30331_df.loc[zip30331_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30315 = zip30315_df.loc[zip30315_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30309 = zip30309_df.loc[zip30309_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30340 = zip30340_df.loc[zip30340_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30308 = zip30308_df.loc[zip30308_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30324 = zip30324_df.loc[zip30324_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30342 = zip30342_df.loc[zip30342_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30318 = zip30318_df.loc[zip30318_df['Maximum Temperature'] > 80,['Maximum Temperature','Accident_ct']]
HiTemp30312.head()
```

```python
# accidents with temperature below 80 
LoTemp30312 = zip30312_df.loc[zip30312_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30339 = zip30339_df.loc[zip30339_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30331 = zip30331_df.loc[zip30331_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30315 = zip30315_df.loc[zip30315_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30309 = zip30309_df.loc[zip30309_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30340 = zip30340_df.loc[zip30340_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30308 = zip30308_df.loc[zip30308_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30324 = zip30324_df.loc[zip30324_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30342 = zip30342_df.loc[zip30342_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30318 = zip30318_df.loc[zip30318_df['Maximum Temperature'] < 80,['Maximum Temperature','Accident_ct']]
LoTemp30312.head()
```

```python
#high temp 30312

(slope, intercept, rvalue, pvalue, stderr) = linregress(HiTemp30312['Maximum Temperature'], HiTemp30312['Accident_ct'])
regress_values = slope * HiTemp30312['Maximum Temperature'] + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(HiTemp30312['Maximum Temperature'], HiTemp30312['Accident_ct'])
plt.plot(HiTemp30312['Maximum Temperature'], regress_values, color = 'red',label = "High Temp fitted line")
plt.legend()
plt.annotate(line_eq,(80,15.2),fontsize=10,color="red")
plt.xlabel("Temperature")
plt.ylabel("Accidents")
plt.title("Accident Data 30312")
corr=round(st.pearsonr(HiTemp30312['Maximum Temperature'], HiTemp30312['Accident_ct'])[0],2)*100
print(f"The correlation between High Temperature and accidents for zip code 30312 is {corr}%")
print ("slope: %f  intercept: %f" % (slope, intercept))
print("R-Squared: %f" % rvalue**2)



#low temp 30312

(slope, intercept, rvalue, pvalue, stderr) = linregress(LoTemp30312['Maximum Temperature'], LoTemp30312['Accident_ct'])
regress_values = slope * LoTemp30312['Maximum Temperature'] + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(LoTemp30312['Maximum Temperature'], LoTemp30312['Accident_ct'])
plt.plot(LoTemp30312['Maximum Temperature'], regress_values, color = "green",label = "Low Temp fitted line")
plt.legend()
plt.annotate(line_eq,(40,12.5),fontsize=10,color="green")
plt.xlabel("Temperature")
plt.ylabel("Accidents")
plt.title("Accident Data 30312")
corr=round(st.pearsonr(LoTemp30312['Maximum Temperature'], LoTemp30312['Accident_ct'])[0],2)*100
print(f"The correlation between Low Temperature and accidents for zip code 30312 is {corr}%")
print ("slope: %f  intercept: %f" % (slope, intercept))
print("R-Squared: %f" % rvalue**2)
```
