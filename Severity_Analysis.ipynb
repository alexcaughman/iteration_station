```python
#import dependencies
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading in weather data from csv
Csvpath = "weather_data2018-19_csv.csv"
weather_data = pd.read_csv(Csvpath)
weather_data['Date time']= pd.to_datetime(weather_data['Date time']) 
weather_data['Date'] = weather_data['Date time'].dt.strftime('%m/%d/%Y')
weather_data = weather_data[['Location', 'Address', 'Country', 'Date', 'Maximum Temperature',
       'Minimum Temperature', 'Temperature', 'Wind Chill', 'Heat Index',
       'Precipitation', 'Snow Depth', 'Wind Speed', 'Wind Gust', 'Cloud Cover',
       'Relative Humidity', 'Conditions']]
weather_data
```

```python
#reading in traffic data from csv Traffic csvs
Csvpath_30308 = "Traffic_Resources/30308_Traffic_data_final.csv"
Csvpath_30309 = "Traffic_Resources/30309_Traffic_data_final.csv"
Csvpath_30312 = "Traffic_Resources/30312_Traffic_data_final.csv"
Csvpath_30315 = "Traffic_Resources/30315_Traffic_data_final.csv"
Csvpath_30318 = "Traffic_Resources/30318_Traffic_data_final.csv"
Csvpath_30324 = "Traffic_Resources/30324_Traffic_data_final.csv"
Csvpath_30331 = "Traffic_Resources/30331_Traffic_data_final.csv"
Csvpath_30339 = "Traffic_Resources/30339_Traffic_data_final.csv"
Csvpath_30340 = "Traffic_Resources/30340_Traffic_data_final.csv"
Csvpath_30342 = "Traffic_Resources/30342_Traffic_data_final.csv"


traffic_data_30312 = pd.read_csv(Csvpath_30312)
traffic_data_30315 = pd.read_csv(Csvpath_30315)
traffic_data_30309 = pd.read_csv(Csvpath_30309)
traffic_data_30339 = pd.read_csv(Csvpath_30339)
traffic_data_30331 = pd.read_csv(Csvpath_30331)
traffic_data_30340 = pd.read_csv(Csvpath_30340)
traffic_data_30324 = pd.read_csv(Csvpath_30324)
traffic_data_30308 = pd.read_csv(Csvpath_30308)
traffic_data_30318 = pd.read_csv(Csvpath_30318)
traffic_data_30342 = pd.read_csv(Csvpath_30342)
traffic_data_30339
```

```python
#creating separate weather dataframes for each zipcode
weather_data_30312 = weather_data.loc[weather_data["Address"]==30312,:]
weather_data_30315 = weather_data.loc[weather_data["Address"]==30315,:]
weather_data_30309 = weather_data.loc[weather_data["Address"]==30309,:]
weather_data_30339 = weather_data.loc[weather_data["Address"]==30339,:]
weather_data_30331 = weather_data.loc[weather_data["Address"]==30331,:]
weather_data_30340 = weather_data.loc[weather_data["Address"]==30340,:]
weather_data_30324 = weather_data.loc[weather_data["Address"]==30324,:]
weather_data_30308 = weather_data.loc[weather_data["Address"]==30308,:]
weather_data_30318 = weather_data.loc[weather_data["Address"]==30318,:]
weather_data_30342 = weather_data.loc[weather_data["Address"]==30342,:]
weather_data_30315
```

```python
#30312 traffic and weather merge
weather_traffic_30312 = pd.merge(weather_data_30312, traffic_data_30312, on = "Date", how = "left")
weather_traffic_30312["Accident_ct"].sum()
```

```python
#Create Severity DataFrame
Severity30312 = {'Severity1': weather_traffic_30312["Severity_1"],
                        'Severity2': weather_traffic_30312["Severity_2"],
                        'Severity3': weather_traffic_30312["Severity_3"],
                        'Severity4': weather_traffic_30312["Severity_4"],
                        'Dates': weather_traffic_30312 ['Date']}
Severity30312_df = pd.DataFrame (Severity30312, columns = ['Severity1','Severity2','Severity3','Severity4','Dates'])

#Replace NaN Values with Zeroes
Severity30312_df.fillna(0, inplace=True)

#Convert 'Dates' to datetime format
Severity30312_df['Dates'] = Severity30312_df['Dates'].astype('datetime64[ns]')

#Sort 'Dates' to month/year
Severity30312_df['Dates'] = Severity30312_df['Dates'].dt.strftime('%Y/%m')
Severity30312_df.sort_values(by=['Dates'])

#Group by 'Dates'

Severity30312_df = Severity30312_df.groupby('Dates').sum()

#Print Severity DataFrame
Severity30312_df
```

```python
Severity30312_df.loc[:,['Severity1','Severity2','Severity3','Severity4']].plot.bar(stacked=True, color=(['g','b','orange','r']), figsize=(30,20))
plt.legend(loc = "upper right", fontsize = 20)
plt.title("Total 30312 Accident Severity Analysis", fontsize = 80)
plt.xlabel("Severity Level by Month", fontsize = 80)
plt.ylabel("Total Accidents", fontsize = 80)
plt.xticks(rotation=45,fontsize =30)
plt.yticks(fontsize =30)
plt.savefig("Images/Total_30312_Accident_Severity_Analysis.png")
```
