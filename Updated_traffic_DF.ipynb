```python
import csv
import pandas as pd
```

```python
#reading in traffic data from csv
Csvpath = "US_Accidents_Dec19.csv"
traffic_data = pd.read_csv(Csvpath)
traffic_data.head()
```

```python
traffic_data['Corrected_Zipcode'] = traffic_data['Zipcode'].str.slice(0,5)
traffic_data
```

```python
#Finding all column names in traffic dataframe
traffic_data.columns
```

```python
#updating dataframe to keep only columns we need
traffic_data_reorg = traffic_data[["Severity", "Description", "Start_Time", "Start_Lat","Start_Lng","Street", "Corrected_Zipcode", "City","County","State","Visibility(mi)"]]
traffic_data_reorg = traffic_data_reorg.rename(columns={"Corrected_Zipcode":"Zipcode"})
traffic_data_reorg.head()
```

```python
#Narrowing down dataframe to only show Atlanta, Georgia accidents from 2018
date = '2018-01-01 00:00:00'
atl_data = traffic_data_reorg.loc[(traffic_data_reorg["State"]=="GA")
                                  &(traffic_data_reorg["City"]=="Atlanta")
                                  &(traffic_data_reorg["Start_Time"]>=date),:]
```

```python
#narrowing down dataframe to only show zipcodes with highest number of accidents
narrowed_atl_df = atl_data.loc[(atl_data["Zipcode"]=='30312')
                               |(atl_data["Zipcode"]=='30339')
                               |(atl_data["Zipcode"]=='30331')
                               |(atl_data["Zipcode"]=='30315')
                               |(atl_data["Zipcode"]=='30309')
                               |(atl_data["Zipcode"]=='30340')
                               |(atl_data["Zipcode"]=='30308')
                               |(atl_data["Zipcode"]=='30324')
                               |(atl_data["Zipcode"]=='30342')
                               |(atl_data["Zipcode"]=='30318'),:]
narrowed_atl_df
```

```python
narrowed_atl_df.dtypes
narrowed_atl_df['Start_Time']= pd.to_datetime(narrowed_atl_df['Start_Time']) 
narrowed_atl_df['Date'] = narrowed_atl_df['Start_Time'].dt.strftime('%m/%d/%Y')
narrowed_atl_df
```

```python
narrowed_atl_df["Street"].value_counts()
```

```python
#https://stackoverflow.com/questions/50474289/in-pandas-how-to-search-for-words-and-phrases-to-create-new-dataframe
str_choice1 = "St|(St NE)|(St NW)|(St SW)|(St SE)|(Dr NE)|(Dr SW)|(Dr SE)|(Dr NW)|Ave|(Ave NE)|\
(Ave SW)|(Ave SE)|(Ave NW)|Rd|(Rd SE)|(Rd SW)|(Rd NW)|(Rd NE)|Blvd|Dr|(Blvd NE)|(Blvd SW)|Way|(Way SE)|\
Park|Boulevard|(Cir NE)|(Cir NW)|Sq|(Pl NE)|(Pl SW)|(Ct NW)|(Ln SW)|(Ln NW)|(Ctr NW)|(Ctr SE)|(Chase NE)|(Trl SW)" 
street_df = narrowed_atl_df[narrowed_atl_df['Street'].str.contains(str_choice1, na = False)]
street_df["Rd_Type"] = "Residential"
street_df
```

```python
str_choice2 = "GA-|I-|Pkwy|Hwy|US-|400|Connector|Expy" 
highway_df = narrowed_atl_df[narrowed_atl_df['Street'].str.contains(str_choice2, na = False)]
highway_df["Rd_Type"] = "Interstate"
highway_df
```

```python
New_traffic_info = pd.concat([street_df, highway_df], ignore_index=True)
New_traffic_info
```

```python
#30312
traffic_info_30312 = pd.DataFrame(index=pd.date_range("01/01/2018", periods=730))
traffic_info_30312.index = traffic_info_30312.index.strftime('%m/%d/%Y')
sev_1 = []
sev_2 = []
sev_3 = []
sev_4 = []
num_accidents = []
Interstate_num = []
Residential = []

for i in range(len(traffic_info_30312)):
    traffic_info_df = New_traffic_info.loc[New_traffic_info["Zipcode"]=="30312",:]
    traffic_info_df = traffic_info_df.loc[traffic_info_df["Date"]==traffic_info_30312.index[i],:]
    sev_1.append(len(traffic_info_df.loc[traffic_info_df["Severity"]==1,:]))
    sev_2.append(len(traffic_info_df.loc[traffic_info_df["Severity"]==2,:]))
    sev_3.append(len(traffic_info_df.loc[traffic_info_df["Severity"]==3,:]))
    sev_4.append(len(traffic_info_df.loc[traffic_info_df["Severity"]==4,:]))
    Interstate_num.append(len(traffic_info_df.loc[traffic_info_df["Rd_Type"]=="Interstate",:]))
    Residential.append(len(traffic_info_df.loc[traffic_info_df["Rd_Type"]=="Residential",:]))
    num_accidents.append(len(traffic_info_df))
traffic_info_30312["Severity_1"] = sev_1
traffic_info_30312["Severity_2"] = sev_2
traffic_info_30312["Severity_3"] = sev_3
traffic_info_30312["Severity_4"] = sev_4
traffic_info_30312["Residential_ct"] = Residential
traffic_info_30312["Interstate_ct"] = Interstate_num
traffic_info_30312["Accident_ct"] = num_accidents
traffic_info_30312["Zipcode"] = "30312"
traffic_info_30312.index.name = 'Date'
traffic_info_30312
```
